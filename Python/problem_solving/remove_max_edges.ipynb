# Solution

```python
class Solution:    
    def maxNumEdgesToRemove(self, n: int, edges: List[List[int]]) -> int:
        alice_uf = UnionFind(n)
        bob_uf = UnionFind(n)

        # Count the number of edges added
        edges_added = 0

        # Sort edges by type so type 3 edges come first
        edges.sort(reverse=True)

        # Process type 3 edges
        for edge_type, u, v in edges:
            if edge_type == 3:
                added_to_alice = alice_uf.union(u-1, v-1)
                added_to_bob = bob_uf.union(u-1, v-1)
                if added_to_alice or added_to_bob:
                    edges_added += 1

        # Process type 1 and type 2 edges
        for edge_type, u, v in edges:
            if edge_type == 1:
                if alice_uf.union(u-1, v-1):
                    edges_added += 1
            elif edge_type == 2:
                if bob_uf.union(u-1, v-1):
                    edges_added += 1

        # Check if both Alice's and Bob's graphs are fully connected
        if all(alice_uf.find(i) == alice_uf.find(0) for i in range(n)) and all(bob_uf.find(i) == bob_uf.find(0) for i in range(n)):
            return len(edges) - edges_added
        else:
            return -1
    
class UnionFind:
    def __init__(self, n):
        self.parent = list(range(n))
        self.rank = [0] * n

    def find(self, x):
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x])  # Path compression
        return self.parent[x]

    def union(self, x, y):
        rootX = self.find(x)
        rootY = self.find(y)
        if rootX != rootY:
            if self.rank[rootX] > self.rank[rootY]:
                self.parent[rootY] = rootX
            elif self.rank[rootX] < self.rank[rootY]:
                self.parent[rootX] = rootY
            else:
                self.parent[rootY] = rootX
                self.rank[rootX] += 1
            return True
        return False


# Example usage
n = 4
edges = [[3,2,3],[1,1,2],[2,3,4]]
sol = Solution()
print(sol.maxNumEdgesToRemove(n, edges))  # Output: 2
```
