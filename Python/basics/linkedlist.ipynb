### Operations to perform on Linked-List ⚙️
- Insert at Beginning
- Insert at End
- Insert at particular position
- Insert a List of values
- Delete from particular position
- Delete by value
- Print Linked-List
- Get length of Linked-List

# Code for Linked List ⛓️‍💥


```python
#-------------------------------------------------------------------DEFINING NODE-------------------------------------------------------------------
class Node:
    def __init__(self, data=None, next=None):
        self.data = data
        self.next = next

#--------------------------------------------------------------CLASS WITH ALL FUNCTIONS-------------------------------------------------------------
class LinkedList:
    def __init__(self):
        self.head = None

#-----------------------------------------------------------------INSERT AT BEGINNING---------------------------------------------------------------
    def insert_at_begining(self, data):
        node = Node(data, self.head)
        self.head = node

#-------------------------------------------------------------------INSERT AT END-------------------------------------------------------------------
    def insert_at_end(self, data):
        node = Node(data)
        if self.head is None:
            self.head = node
            return
        current_node = self.head
        while current_node.next:
            current_node = current_node.next

        current_node.next = node

#--------------------------------------------------------------------INSERT LIST--------------------------------------------------------------------
    def insert_values(self, data_list):
        self.head = None
        for data in data_list:
            # node = Node(data, self.head)
            # self.head = node
            self.insert_at_end(data)

#-----------------------------------------------------------------PRINT LINKED-LIST-----------------------------------------------------------------        
    def print_ll(self):
        if self.head is None:
            print("Empty Linked List")
        current_node = self.head
        listr = ''
        while current_node:
            listr += str(current_node.data) + ' --> '
            # print(current_node.data)
            current_node = current_node.next
        print(listr)

#---------------------------------------------------------------LENGTH OF LINKED-LIST---------------------------------------------------------------
    def length_of_linkedlist(self):
        counter = 0
        current_node = self.head
        while current_node:
            counter += 1
            current_node = current_node.next
        return counter

#------------------------------------------------------------DELETE FROM PARTICULAR INDEX-----------------------------------------------------------
    def remove_at(self, index):
        if index<0 or index>self.length_of_linkedlist():
            raise Exception("Invalid Index")
        if index == 0:
            self.head = self.head.next
            return
        counter = 0
        current_node = self.head
        while current_node:
            if counter == index-1:
                current_node.next = current_node.next.next
                break
            current_node = current_node.next
            counter += 1

#------------------------------------------------------------INSERT AT PARTICULAR INDEX-------------------------------------------------------------
    def insert_at(self, index, data):
        if index<0 or index>self.length_of_linkedlist():
            raise Exception("Invalid Index")
        if index == 0:
            self.insert_at_begining()
            return
        counter = 0
        current_node = self.head
        while current_node:
            if counter == index-1:
                node = Node(data, current_node.next)
                current_node.next = node
                break
            current_node = current_node.next
            counter += 1

#-----------------------------------------------------------------DELETE BY VALUE-------------------------------------------------------------------
    def remove_by_value(self, data):
     current_node = self.head
     counter = 0
     if current_node.data == data:
          self.head = current_node.next
          return
     while current_node:
          if current_node.next.data == data:
               current_node.next = current_node.next.next
               break
          current_node = current_node.next
          counter += 1
```
